```python
import sentencepiece as spm

# Load the trained SentencePiece model
sp = spm.SentencePieceProcessor()
sp.load("mundari_spm.model")
```

```python
# Tokenize a sample sentence
text = "इनिआः काजि मेनाः चि टुङ गालाङ राः"
tokens = sp.encode_as_pieces(text)
print("Tokens:", tokens)
```

```python
# Detokenize the tokens back to text
detokenized_text = sp.decode_pieces(tokens)
print("Detokenized:", detokenized_text)
```

```python
print("Vocabulary Size:", sp.get_piece_size())
```

```python
# List the first 5 tokens in the vocabulary
for i in range(5):
    print(f"Token {i}: {sp.id_to_piece(i)}")
```

```python
# Convert tokens to IDs
token_ids = sp.encode_as_ids(text)
print("Token IDs:", token_ids)

# Convert IDs back to tokens
tokens_from_ids = [sp.id_to_piece(id_) for id_ in token_ids]
print("Tokens from IDs:", tokens_from_ids)

# Convert IDs back to text
decoded_text = sp.decode_ids(token_ids)
print("Decoded Text:", decoded_text)
```

```python
%%bash
echo "Hardcoding SentencePiece Tokenization in Colab"

# Hardcoded paths for directories
exp_dir="/content/IndicTrans2"
data_dir="/content/IndicTrans2/data"
bpe_dir="/content/IndicTrans2/bpe"

# Create output directory for BPE tokenized data
mkdir -p $bpe_dir

# Hardcoded language settings and split names
src_lang="mun_Deva"
tgt_lang="hin_Deva"
splits=("train" "dev")

# Hardcode parallel installation status
parallel_installed=false

# Tokenize each split
for split in "${splits[@]}"; do
    in_split_dir="$data_dir/$split"
    out_split_dir="$bpe_dir/$split"

    echo "Apply SentencePiece tokenization to SRC corpus for $split"
    if $parallel_installed; then
        parallel --pipe --keep-order \
            spm_encode --model=$exp_dir/vocab/model.SRC.model \
                       --output_format=piece \
                       < $in_split_dir.$src_lang \
                       > $out_split_dir.$src_lang
    else
        spm_encode --model=$exp_dir/vocab/model.SRC.model \
                   --output_format=piece \
                   < $in_split_dir.$src_lang \
                   > $out_split_dir.$src_lang
    fi

    echo "Apply SentencePiece tokenization to TGT corpus for $split"
    if $parallel_installed; then
        parallel --pipe --keep-order \
            spm_encode --model=$exp_dir/vocab/model.TGT.model \
                       --output_format=piece \
                       < $in_split_dir.$tgt_lang \
                       > $out_split_dir.$tgt_lang
    else
        spm_encode --model=$exp_dir/vocab/model.TGT.model \
                   --output_format=piece \
                   < $in_split_dir.$tgt_lang \
                   > $out_split_dir.$tgt_lang
    fi
done

echo "SentencePiece Tokenization Complete"
```

```python
%%bash
echo "Binarizing data for Fairseq"

# Hardcoded paths for directories
exp_dir="/content/IndicTrans2"
data_dir="$exp_dir/final"
out_data_dir="$exp_dir/final_bin"

# Use CPU count to determine the number of workers
num_workers=$(python3 -c "import multiprocessing; print(multiprocessing.cpu_count())")

# Create the output directory for binarized data
mkdir -p $out_data_dir

# Run Fairseq preprocess
fairseq-preprocess \
    --source-lang SRC --target-lang TGT \
    --trainpref $data_dir/train \
    --validpref $data_dir/dev \
    --destdir $out_data_dir \
    --workers $num_workers \
    --srcdict $exp_dir/vocab/model.SRC.vocab \
    --tgtdict $exp_dir/vocab/model.TGT.vocab

echo "Binarization complete"
```
